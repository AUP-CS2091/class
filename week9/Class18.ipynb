```python
def simulate_means(size):
    means = make_array()
    for _ in np.arange(1000):
        flights_sample = flights.sample(size)
        flights_mean = np.mean(flights_sample.column(3))
        means = np.append(means, flights_mean)
    return means
    
```

```python
std_1000 = np.std(simulate_means(1000))
std_1000
```

```python
stds = Table() \
    .with_column("1000", std_1000) \
    .with_column("700", np.std(simulate_means(700))) \
    .with_column("500", np.std(simulate_means(500))) \
    .with_column("300", np.std(simulate_means(300))) \
    .with_column("100", np.std(simulate_means(100))) 

stds
```
